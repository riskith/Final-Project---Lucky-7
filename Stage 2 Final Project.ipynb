### Handling Outlier

```python
# Menghapus outlier pada kolom ADR

Q1 = df_droped['adr'].quantile(0.25)
Q3 = df_droped['adr'].quantile(0.75)
IQR = Q3-Q1
low_limit = Q1 - (1.5*IQR)
high_limit = Q3 + (1.5*IQR)
filtered = ((df_droped['adr'] >= low_limit) & (df_droped['adr'] <= high_limit))
df_clear = df_droped[filtered]
```

```python
### numeric pipe

nums =['lead_time','arrival_date_week_number','arrival_date_day_of_month','stays_in_weekend_nights','stays_in_week_nights','adults','children','babies','is_repeated_guest','previous_cancellations','previous_bookings_not_canceled','booking_changes','days_in_waiting_list','adr','required_car_parking_spaces','total_of_special_requests','total_hari']
scaler= StandardScaler()
transformer= PowerTransformer(method= 'yeo-johnson')
num_pipe= Pipeline([
('scaler', scaler), 
('transformer', transformer)
])

# ordinal
ordinal_var = ['assigned_room_type', 'reserved_room_type']
ordinal_encoder = OrdinalEncoder(categories=[ordinal_var])
ordinal_pipe = Pipeline([
    ('ordinal_encoder', ordinal_encoder)
])

#nominal(Binary)
nominal_binary_var = ['hotel', 'is_repeated_guest']
nominal_binary_encoder = ce.BinaryEncoder()
nominal_binary_pipe = Pipeline([
    ('nominal_binary_encoder', nominal_binary_encoder)
])

# nominal (Onehot)
nominal_var = ['arrival_date_month', 'country', 'market_segment', 'distribution_channel', 'customer_type', 'reservation_status', 'reservation_years', 'reservation_month', 'reservation_dateday']
nominal_encoder = OneHotEncoder(handle_unknown='ignore')
nominal_pipe = Pipeline([
    ('nominal_encoder', nominal_encoder)
])
```

```python
preprocessor = ColumnTransformer([
    ('num_pipe', num_pipe, nums),
    ('ordinal_pipe', ordinal_pipe, ordinal_var),
    ('nominal_pipe', nominal_pipe, nominal_var),
    ('nominal_binary_pipe', nominal_binary_pipe, nominal_binary_var)
])
```

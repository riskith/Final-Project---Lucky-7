```python
classification_list = []

for index, column in df.iterrows():
    if column['total_hari'] <1:
        classification = '<1'
    elif column['total_hari']==1:
        classification = '1'
    elif column['total_hari'] >= 2 and column['total_hari'] <=6:
        classification = '2-6'
    elif column['total_hari'] >= 7 and column['total_hari'] <=11:
        classification = '7-11'
    elif column['total_hari'] >= 12 and column['total_hari'] <=16:
        classification = '>12'
    else:
        classification = '>17'
    
    classification_list.append(classification)
    
df['Classification'] = classification_list
df
```

```python
sns.countplot(data = df, x = 'Classification',hue = 'is_canceled')
plt.legend(loc=(.65,.8))
```
